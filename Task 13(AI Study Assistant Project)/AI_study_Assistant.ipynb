```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.chains import RetrievalQA
from transformers import pipeline
import os
from google.colab import files
```

```python
loader = TextLoader("AI_Study_Assistant_Knowledge_Base.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
def get_answer(query):
    response = qa_pipeline(question=query, context=" ".join([doc.page_content for doc in docs]))
    return response['answer']

    while True:
      query = input("Ask a question (or type 'exit'): ")
      if query.lower() == 'exit':
          break
      answer = get_answer(query)
      print("Answer:", answer)
```

```python
from transformers import pipeline

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=-1)

# Your long text (put the full context here)
context = """
AI Study Assistant Knowledge Base

1. What is Artificial Intelligence?
Artificial Intelligence (AI) is the branch of computer science that aims to create machines that can perform tasks that typically require human intelligence.

2. Types of AI:
- Narrow AI: AI systems that are trained to perform a single task (e.g., voice assistants).
- General AI: A type of AI that can perform any intellectual task a human can do.
- Super AI: Hypothetical AI that surpasses human intelligence in all aspects.

3. Applications of AI:
- Natural Language Processing (NLP)
- Image Recognition
- Autonomous Vehicles
- Robotics
- Expert Systems

...
"""

# Split if text is too long (BART has a ~1024 token limit)
chunks = [context[i:i+1024] for i in range(0, len(context), 1024)]
summary = ""

for chunk in chunks:
    result = summarizer(chunk, max_length=150, min_length=40, do_sample=False)
    summary += result[0]['summary_text'] + " "

print("Summary:", summary.strip())
```

```python
from transformers import pipeline
import gradio as gr

# Load the generative QA model
qa_model = pipeline("text2text-generation", model="google/flan-t5-large", device=-1)

# Your context (longer context is OK with Flan-T5)
context = """
Artificial Intelligence is a field of computer science that focuses on creating smart machines capable of performing tasks that typically require human intelligence.
It includes machine learning, deep learning, computer vision, and natural language processing. The goal of AI is to enable computers to perform intellectual tasks like problem solving, decision making, and understanding language.
"""

def answer_question(question):
    # Create prompt for generative QA
    prompt = f"Question: {question}\nContext: {context}"
    result = qa_model(prompt, max_length=300, do_sample=False)
    return result[0]["generated_text"]

# Gradio interface
gr.Interface(
    fn=answer_question,
    inputs="text",
    outputs="text",
    title="AI Study Assistant 🤖📚",
    description="Apna sawal poochein aur AI se tafseel se jawab paayen!"
).launch(share=True)
```
